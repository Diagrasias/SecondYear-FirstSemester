```python
#4.3.1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp

num = 500
arr = np.random.uniform(0, 1, num)

mean = np.mean(arr)
median = np.median(arr)

result = ttest_1samp(arr, popmean=median, alternative='greater')

if result.pvalue < 0.05:
    print("Average is significantly greater than median")
else:
    print("No significant difference between average and median")

plt.scatter(arr, arr)
plt.show()
```

```python
#4.3.2
import numpy as np
import matplotlib.pyplot as plt

# Задаем функцию
def f(x):
    return (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2))) / (np.abs(1 - np.sin(x)**3)) + np.log(np.abs(2*x))

# Создаем массив x от 1 до 10
x = np.linspace(1, 10, 10)

# Создаем массив значений функции
y = f(x)

# Создаем срез первой половины массива
y_sliced = y[:5]
x_sliced = x[:5]

# Построение графиков
plt.figure(figsize=(10, 5))

# Основной график - линейный
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-', label='Основной массив')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График основного массива')
plt.legend()

# График для среза - точечный
plt.subplot(1, 2, 2)
plt.scatter(x_sliced, y_sliced, c='r', label='Срез первой половины')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Точечный график среза первой половины')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#4.3.3
from scipy.integrate import simps
from numpy import trapz
import numpy as np
import matplotlib.pyplot as plt

# Задаем функцию
def f(x):
    return np.abs(np.cos(x*np.exp(np.cos(x)+np.log(x+1))))

# Создаем массив x от 0 до 10 с шагом 1
x = np.arange(0, 11, 1)

# Создаем массив значений функции
y = f(x)

# Построение графика с заливкой площади
plt.plot(x, y, 'b-', label='f(x)')
plt.fill_between(x, y, 0, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции с заливкой площади')

# Вычисление площади под графиком
area_trapezoidal = trapz(y, x)
area_simps = simps(y, x)
print("Площадь (правило трапеции):", area_trapezoidal)
print("Площадь (правило Симпсона):", area_simps)

plt.legend()
plt.show()
```

```python
#4.3.5
import numpy as np

def calculator(x, y, operator):
    if operator == '+':
        return x + y
    elif operator == '-':
        return x - y
    elif operator == '*':
        return x * y
    elif operator == '/':
        return x / y
    elif operator == 'e^(x+y)':
        return np.exp(x + y)
    elif operator == 'sin(x+y)':
        return np.sin(x + y)
    elif operator == 'cos(x+y)':
        return np.cos(x + y)
    elif operator == 'x^y':
        return np.power(x, y)
    else:
        return "Неверный оператор"
        
# Пример использования
x = 2
y = 3
operators = ['+', '-', '*', '/', 'e^(x+y)', 'sin(x+y)', 'cos(x+y)', 'x^y']

for operator in operators:
    result = calculator(x, y, operator)
    print(f"Результат вычисления {operator}: {result}")
```
